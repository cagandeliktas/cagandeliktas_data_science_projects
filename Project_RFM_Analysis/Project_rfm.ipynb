### Preparing the rfm data frame

```python
rfm = pd.DataFrame({"master_id": df["master_id"],
                    "recency": (analysis_date - df["last_order_date"]).astype('timedelta64[ns]'),
                    "frequency": df["total_order_num"],
                    "monetary": df["total_customer_value"]})
rfm.head()
```

```python
rfm["recency_score"] = pd.qcut(rfm["recency"], q = 5, labels=[5,4,3,2,1])
rfm["frequency_score"] = pd.qcut(rfm["frequency"].rank(method="first"), q = 5, labels=[1,2,3,4,5])
rfm["monetary_score"] = pd.qcut(rfm["monetary"], q = 5, labels=[1,2,3,4,5])

rfm.head()
```

```python
rfm["RF_Score"] = rfm.recency_score.astype("str") + rfm.frequency_score.astype("str")
rfm.head()
```

```python
seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'
}

# AdÄ±m 2
rfm['segment'] = rfm['RF_Score'].replace(seg_map, regex=True)
rfm.head()
```

### customers in "champions" segment have low recency, high frequency and high monetary. these customers are valuable to the company

```python
rfm.groupby("segment").aggregate({"recency": "mean",
                                  "frequency": "mean",
                                  "monetary":"mean"}).sort_values(by = ["frequency", "monetary"], ascending = False)
```
